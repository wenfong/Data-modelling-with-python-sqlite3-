```python
import pandas as pd
```

```python
data = pd.read_csv('GTA_vehicle.csv')
df = pd.DataFrame(data)
df.head()
```

The header row is useless. Let's get rid of it.

```python
df.columns = df.iloc[0]
df = df[1:]
df.tail()
```

The last three rows are also not needed.

```python
df = df[:491]
df
```

There's also unecessary rows for each new vehicle class. Those can be removed as well.

```python
df = df[df['Vehicle_Name'].notna()]
```

The columns 'Sell_Price', 'Stealable', 'Avg w/o Brake', 'Avg w/o Accel', 'Avg w/o Speed', 'Overall','Picture','Notes', and 'Based on' can be removed as they are not needed.

```python
df = df.drop(df.columns[[3,4,10,11,12,13,14,15,16]], axis=1)

df.dtypes
```

The data types of all the columns need to be changed except for Vehichle_Class and Vehicle_Name. Starting with Buy_Price and from looking at the previous outputs, there are unwanted characters that will mess up the conversion to floats. More specifically, the commas, dollar signs as well as invalid values need to be removed.

```python
df['Buy_Price'] = df['Buy_Price'].apply(lambda x: str(x).replace(',','')[1:])
df.loc[df['Buy_Price']=='an','Buy_Price'] = None   
df.loc[df['Buy_Price']=='','Buy_Price'] = None
df['Buy_Price']
```

Now that the characters have been removed, the data type for Buy_Price can be converted to float, along with the data types for Speed, Accel, Brake, and Handl. Capacity is changed to int.

```python
df[['Buy_Price','Speed','Accel','Brake','Handl']] = df[['Buy_Price','Speed','Accel','Brake','Handl']].astype(float)    
df['Capacity'] = df['Capacity'].astype(int)  
```

Missing values from Buy_Price can be replaced with the median values of their respective class.

```python
df['Buy_Price'] = df.groupby('Vehicle_Class')['Buy_Price'].apply(lambda x: x.fillna(x.median()))
```

A new column 'Vehicle_id', a unique ID for each vehicle, is generated by taking three letters based on Vehicle_Class followed by a 3 digit number for each vehicle in its class.

```python
def label_vehicle_id(row, classes, class_labels):

    for label in class_labels:
        if row[0] == label:
            ret_str = class_labels[label]+str(classes[row[0]]).zfill(3)
            classes[row[0]] += 1
            return ret_str 

classes = { i:0 for i in df.Vehicle_Class.unique() }
class_labels = {
    'SUPER':'SUP', 'SPORTS':'SPO', 'SPORTS CLASSIC':'SPC', 'MUSCLE':'MUS', 
    'SEDANS':'SED', 'COUPES':'COU', 'COMPACT':'CMP', 'SUVs':'SUV', 'OFF-ROAD':'OFR',
    'MOTORCYCLES':'MOT', 'GANG':'GAN', 'VANS':'VAN', 'EMERGENCY':'EME', 
    'SERVICE':'SER', 'MILITARY':'MIL', 'COMMERCIAL':'COM', 'INDUSTRIAL':'IND',
    'HELICOPTERS':'HEL', 'PLANES':'PLA', 'BOATS':'BOA', 'CYCLES':'CYC'}

df['Vehicle_id'] = df.apply(lambda row: label_vehicle_id(row, classes, class_labels), axis=1)   
df
```

The data is now ready to be loaded into the database for use.
